```python
def rename(name):
  if name=="Royal Challengers Bangalore":
    return "RCB"
  elif name=="Chennai Super Kings":
    return "CSK"
  else:
    return name
```

```python
rename("Chennai Super Kings")
```

```python
#convert name is given function in a particulor row
pd.read_csv("IPL Matches 2008-2020.csv",converters={"team1":rename,"team2":rename})
```
